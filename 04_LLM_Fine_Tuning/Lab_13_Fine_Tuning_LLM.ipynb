## Part 2: Teaching Style Definition

### My Custom Teaching Style: "Analogy-Based Step-by-Step"

I define my teaching style as a combination of **real-life analogies** and **structured step-by-step breakdowns**. This style is designed specifically for beginner to intermediate computer science students who may struggle with abstract concepts.

#### Structure of every explanation:
1. **Analogy Hook** — Start with a familiar, everyday scenario that mirrors the concept
2. **Step-by-Step Breakdown** — Decompose the concept into 4 clear, numbered steps
3. **Key Takeaway** — End with a one-sentence summary that reinforces the core idea

#### Why this style works:
- **Analogies** bridge the gap between abstract CS concepts and concrete understanding
- **Step-by-step structure** prevents cognitive overload by breaking complex topics into digestible pieces
- **Consistent format** helps students know what to expect, reducing anxiety when studying new topics
- Research in educational psychology shows that **analogical reasoning** significantly improves transfer learning

#### How the dataset follows this style:
Every single sample in my dataset strictly follows this format: analogy first, four numbered steps, then a key takeaway. This consistency is critical for the fine-tuned model to learn and replicate the pattern.

## Part 4: Fine-Tuning Strategy — LoRA (Low-Rank Adaptation)

### Why LoRA:

| Aspect | Details |
|--------|---------|
| **Method** | LoRA (Low-Rank Adaptation) |
| **Library** | `peft` by Hugging Face |
| **Trainable params** | ~0.3M out of 82M total (~0.4%) |
| **Target modules** | `c_attn`, `c_proj` (attention layers) |

### Why LoRA is suitable for this task:

1. **Small dataset (25 samples)**: Full fine-tuning on 25 samples would cause severe overfitting. LoRA limits the number of trainable parameters, acting as a natural regularizer.

2. **Preserves pre-trained knowledge**: LoRA freezes the original model weights and adds small trainable matrices. This means the model keeps its general language ability while learning our teaching style.

3. **Memory efficient**: Only ~0.4% of parameters are updated, so training fits easily in Colab free GPU memory.

4. **Fast training**: With so few trainable parameters, each epoch completes in seconds.

### How LoRA works (conceptually):
- Instead of updating a full weight matrix W (dimensions d × d), LoRA decomposes the update into two small matrices: A (d × r) and B (r × d), where r << d
- The effective weight becomes W + A × B
- Only A and B are trained; W stays frozen
- This dramatically reduces trainable parameters while preserving model quality

### LoRA configuration I am using:
- `r=16` — rank of the adaptation matrices (controls capacity)
- `lora_alpha=32` — scaling factor (higher = stronger adaptation)
- `lora_dropout=0.05` — dropout for regularization
- Target modules: `c_attn` and `c_proj` — the attention weight matrices in GPT-2

````python
# ============================================================
# Load Base Model and Tokenizer
# ============================================================
MODEL_NAME = "distilgpt2"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
tokenizer.pad_token = tokenizer.eos_token

base_model = AutoModelForCausalLM.from_pretrained(MODEL_NAME).to(device)
print(f"Base model loaded: {MODEL_NAME}")
print(f"Total parameters: {sum(p.numel() for p in base_model.parameters()):,}")
````

````python
# ============================================================
# PART 5 (Baseline): Generate output from BASE model BEFORE fine-tuning
# ============================================================
def generate_response(model, prompt, max_new_tokens=250):
    model.eval()
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            temperature=0.7,
            top_p=0.9,
            do_sample=True,
            repetition_penalty=1.2,
            pad_token_id=tokenizer.eos_token_id,
        )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Test prompts
test_prompts = [
    "Explain dynamic programming",
    "Explain recursion using a real-life analogy.",
    "Explain how a stack works",
]

print("=" * 70)
print("BASE MODEL OUTPUTS (Before Fine-Tuning)")
print("=" * 70)

base_outputs = {}
for prompt in test_prompts:
    output = generate_response(base_model, prompt)
    base_outputs[prompt] = output
    print(f"\nPrompt: {prompt}")
    print(f"Output: {output}")
    print("-" * 70)
````

````python
# ============================================================
# Prepare Dataset for Training (with Label Masking)
# ============================================================
# Label masking ensures the model only learns to generate RESPONSES,
# not memorize the instruction prompts. Instruction tokens are set
# to -100 so they are ignored during loss computation.

class CSTeachingDataset(Dataset):
    def __init__(self, samples, tokenizer, max_length=512):
        self.encodings = []
        for s in samples:
            instruction_part = f"### Instruction:\n{s['instruction']}\n\n### Response:\n"
            full_text = f"{instruction_part}{s['output']}{tokenizer.eos_token}"
            encoding = tokenizer(
                full_text,
                truncation=True,
                max_length=max_length,
                padding="max_length",
                return_tensors="pt",
            )
            # Create labels with instruction tokens masked
            labels = encoding["input_ids"].squeeze().clone()
            inst_len = len(tokenizer(instruction_part, add_special_tokens=False)["input_ids"])
            labels[:inst_len] = -100   # Mask instruction tokens
            labels[encoding["attention_mask"].squeeze() == 0] = -100  # Mask padding

            self.encodings.append({
                "input_ids": encoding["input_ids"].squeeze(),
                "attention_mask": encoding["attention_mask"].squeeze(),
                "labels": labels,
            })

    def __len__(self):
        return len(self.encodings)

    def __getitem__(self, idx):
        return self.encodings[idx]

# Split into train (20) and validation (5) sets
train_samples = dataset_samples[:20]
val_samples = dataset_samples[20:]

train_dataset = CSTeachingDataset(train_samples, tokenizer)
val_dataset = CSTeachingDataset(val_samples, tokenizer)
print(f"Training samples: {len(train_dataset)}, Validation samples: {len(val_dataset)}")
print(f"Sample token length: {train_dataset[0]['input_ids'].shape[0]}")

# Show label masking in action
sample_labels = train_dataset[0]['labels']
masked_count = (sample_labels == -100).sum().item()
total_count = len(sample_labels)
print(f"\nLabel masking: {masked_count}/{total_count} tokens masked (instruction + padding)")
print(f"Model only learns from {total_count - masked_count} response tokens per sample")
````

````python
# ============================================================
# Apply LoRA Configuration
# ============================================================
lora_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    target_modules=["c_attn", "c_proj"],
    bias="none",
)

model = get_peft_model(base_model, lora_config)

trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
total_params = sum(p.numel() for p in model.parameters())
print(f"Trainable parameters: {trainable_params:,} ({100 * trainable_params / total_params:.2f}%)")
print(f"Total parameters: {total_params:,}")
print(f"Frozen parameters: {total_params - trainable_params:,}")
````

````python
# ============================================================
# Training Configuration and Fine-Tuning
# ============================================================
training_args = TrainingArguments(
    output_dir="./lora_cs_assistant",
    num_train_epochs=20,
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    learning_rate=2e-4,
    warmup_steps=10,
    weight_decay=0.01,
    logging_steps=5,
    save_strategy="epoch",
    eval_strategy="epoch",          # NEW: evaluate every epoch
    save_total_limit=2,
    fp16=torch.cuda.is_available(),
    report_to="none",
    lr_scheduler_type="cosine",
    seed=42,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,       # NEW: validation set for monitoring
    data_collator=DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False),
)

print("Starting LoRA fine-tuning...")
print(f"Epochs: {training_args.num_train_epochs}")
print(f"Batch size: {training_args.per_device_train_batch_size}")
print(f"Gradient accumulation: {training_args.gradient_accumulation_steps}")
print(f"Effective batch size: {training_args.per_device_train_batch_size * training_args.gradient_accumulation_steps}")
print(f"Learning rate: {training_args.learning_rate}")

train_result = trainer.train()

print(f"\nTraining complete!")
print(f"Final training loss: {train_result.training_loss:.4f}")
````

## Part 5B: Quantitative Evaluation Metrics

Beyond qualitative comparison, I now evaluate the model using **quantitative metrics**:
1. **Perplexity** — measures how "surprised" the model is by the text (lower = better)
2. **BLEU Score** — measures n-gram overlap between generated and reference text
3. **ROUGE-L Score** — measures longest common subsequence between generated and reference text

These metrics provide objective, reproducible evidence of improvement.

````python
# ============================================================
# Quantitative Evaluation: Perplexity Scores
# ============================================================
import math

def calculate_perplexity(eval_model, text, tokenizer, device):
    """Calculate perplexity of a given text under a model.
    Lower perplexity = model is more confident about generating this text."""
    eval_model.eval()
    encodings = tokenizer(text, return_tensors="pt", truncation=True, max_length=512).to(device)
    with torch.no_grad():
        outputs = eval_model(**encodings, labels=encodings["input_ids"])
    return math.exp(outputs.loss.item())

# Evaluate on reference texts from our dataset
eval_texts = [
    f"### Instruction:\n{dataset_samples[0]['instruction']}\n\n### Response:\n{dataset_samples[0]['output'][:200]}",
    f"### Instruction:\n{dataset_samples[2]['instruction']}\n\n### Response:\n{dataset_samples[2]['output'][:200]}",
    f"### Instruction:\n{dataset_samples[5]['instruction']}\n\n### Response:\n{dataset_samples[5]['output'][:200]}",
    f"### Instruction:\n{dataset_samples[8]['instruction']}\n\n### Response:\n{dataset_samples[8]['output'][:200]}",
]

print("=" * 70)
print("PERPLEXITY COMPARISON: Base Model vs Fine-Tuned Model")
print("=" * 70)
print(f"{'Topic':<40} {'Base Model':>12} {'Fine-Tuned':>12} {'Change':>10}")
print("-" * 74)

base_ppls, ft_ppls = [], []
for text in eval_texts:
    topic = text.split('\n')[1][:38]
    base_ppl = calculate_perplexity(base_model, text, tokenizer, device)
    ft_ppl = calculate_perplexity(model, text, tokenizer, device)
    base_ppls.append(base_ppl)
    ft_ppls.append(ft_ppl)
    change = ((ft_ppl - base_ppl) / base_ppl) * 100
    print(f"{topic:<40} {base_ppl:>12.2f} {ft_ppl:>12.2f} {change:>+9.1f}%")

print("-" * 74)
avg_base = sum(base_ppls) / len(base_ppls)
avg_ft = sum(ft_ppls) / len(ft_ppls)
print(f"{'AVERAGE':<40} {avg_base:>12.2f} {avg_ft:>12.2f} {((avg_ft-avg_base)/avg_base)*100:>+9.1f}%")
print(f"\nLower perplexity = model is more confident about the text")
print(f"Fine-tuning {'reduced' if avg_ft < avg_base else 'increased'} average perplexity by {abs(avg_base - avg_ft):.2f}")
````

````python
# ============================================================
# Quantitative Evaluation: BLEU & ROUGE-L Scores
# ============================================================
from collections import Counter

def compute_bleu(reference, hypothesis, max_n=4):
    """Compute BLEU score between reference and hypothesis text."""
    ref_tokens = reference.lower().split()
    hyp_tokens = hypothesis.lower().split()
    if len(hyp_tokens) == 0:
        return 0.0
    scores = []
    for n in range(1, max_n + 1):
        ref_ngrams = Counter([tuple(ref_tokens[i:i+n]) for i in range(len(ref_tokens)-n+1)])
        hyp_ngrams = Counter([tuple(hyp_tokens[i:i+n]) for i in range(len(hyp_tokens)-n+1)])
        clipped = sum(min(hyp_ngrams[ng], ref_ngrams[ng]) for ng in hyp_ngrams)
        total = sum(hyp_ngrams.values())
        scores.append(clipped / total if total > 0 else 0)
    bp = min(1.0, math.exp(1 - len(ref_tokens) / max(len(hyp_tokens), 1)))
    if all(s > 0 for s in scores):
        bleu = bp * math.exp(sum(math.log(s) for s in scores) / len(scores))
    else:
        bleu = 0.0
    return bleu

def lcs_length(x, y):
    """Compute Longest Common Subsequence length."""
    m, n = len(x), len(y)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if x[i-1] == y[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    return dp[m][n]

def compute_rouge_l(reference, hypothesis):
    """Compute ROUGE-L F1 score."""
    ref_tokens = reference.lower().split()
    hyp_tokens = hypothesis.lower().split()
    if len(ref_tokens) == 0 or len(hyp_tokens) == 0:
        return 0.0
    lcs = lcs_length(ref_tokens, hyp_tokens)
    precision = lcs / len(hyp_tokens)
    recall = lcs / len(ref_tokens)
    if precision + recall == 0:
        return 0.0
    return 2 * precision * recall / (precision + recall)

# Evaluate: generate responses and compare with references
eval_pairs = [
    {"prompt": dataset_samples[0]["instruction"], "reference": dataset_samples[0]["output"]},
    {"prompt": dataset_samples[2]["instruction"], "reference": dataset_samples[2]["output"]},
    {"prompt": dataset_samples[5]["instruction"], "reference": dataset_samples[5]["output"]},
    {"prompt": dataset_samples[8]["instruction"], "reference": dataset_samples[8]["output"]},
    {"prompt": dataset_samples[10]["instruction"], "reference": dataset_samples[10]["output"]},
]

print("=" * 70)
print("BLEU & ROUGE-L SCORES (Fine-Tuned Model vs Reference)")
print("=" * 70)

bleu_scores, rouge_scores = [], []
for pair in eval_pairs:
    formatted = f"### Instruction:\n{pair['prompt']}\n\n### Response:\n"
    output = generate_response(model, formatted, max_new_tokens=250)
    if "### Response:" in output:
        output = output.split("### Response:")[-1].strip()
    bleu = compute_bleu(pair["reference"], output)
    rouge_l = compute_rouge_l(pair["reference"], output)
    bleu_scores.append(bleu)
    rouge_scores.append(rouge_l)
    print(f"\nPrompt: {pair['prompt'][:60]}")
    print(f"  BLEU:    {bleu:.4f}")
    print(f"  ROUGE-L: {rouge_l:.4f}")

print(f"\n{'='*70}")
print(f"Average BLEU:    {sum(bleu_scores)/len(bleu_scores):.4f}")
print(f"Average ROUGE-L: {sum(rouge_scores)/len(rouge_scores):.4f}")
print(f"\nNote: Scores below 0.3 are expected — the model rephrases, not copies.")
print(f"Any positive BLEU/ROUGE indicates the model learned relevant vocabulary and structure.")
````

## Part 5C: Generation Parameter Comparison

Different **temperature** and **top-p** values significantly affect generation quality:
- **Low temperature (0.3)**: More focused and deterministic output
- **High temperature (1.2)**: More creative but potentially incoherent
- **Low top-p (0.5)**: Narrow vocabulary, safer word choices
- **High top-p (0.99)**: Wider vocabulary, more diverse phrasing

````python
# ============================================================
# Temperature & Top-p Parameter Comparison
# ============================================================
def generate_with_params(gen_model, prompt, temperature, top_p, max_new_tokens=200):
    """Generate text with specific temperature and top_p settings."""
    gen_model.eval()
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = gen_model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            temperature=temperature,
            top_p=top_p,
            do_sample=True,
            repetition_penalty=1.2,
            pad_token_id=tokenizer.eos_token_id,
        )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

test_prompt = "### Instruction:\nExplain how a queue data structure works using a real-life analogy.\n\n### Response:\n"

param_configs = [
    {"temperature": 0.3, "top_p": 0.9,  "label": "Low Temp (0.3) — More focused/deterministic"},
    {"temperature": 0.7, "top_p": 0.9,  "label": "Medium Temp (0.7) — Balanced (used in training)"},
    {"temperature": 1.2, "top_p": 0.95, "label": "High Temp (1.2) — More creative/random"},
    {"temperature": 0.7, "top_p": 0.5,  "label": "Low Top-p (0.5) — Narrow vocabulary"},
    {"temperature": 0.7, "top_p": 0.99, "label": "High Top-p (0.99) — Wider vocabulary"},
]

print("=" * 80)
print("GENERATION PARAMETER COMPARISON")
print("Prompt: Explain how a queue data structure works using a real-life analogy.")
print("=" * 80)

for config in param_configs:
    output = generate_with_params(model, test_prompt, config["temperature"], config["top_p"])
    if "### Response:" in output:
        output = output.split("### Response:")[-1].strip()
    print(f"\n--- {config['label']} ---")
    print(output[:400])
    print("-" * 80)
````

## Reflection Report

### What the Model Learned
The fine-tuned DistilGPT2 model successfully learned several things from just 25 training samples:
1. **Instruction-following format**: The model learned to recognize the "Instruction → Response" pattern and generate appropriate answers instead of random text continuations
2. **Teaching style**: The analogy-based step-by-step format was partially captured — outputs tend to include analogies and numbered steps
3. **Domain vocabulary**: The model became more likely to use CS-specific terminology correctly within the context of explanations
4. **Structure**: Outputs became more organized with a beginning, middle, and end rather than trailing off

### Where It Failed
1. **Out-of-distribution topics**: When asked about concepts not in the training set (CAP theorem, garbage collection), the model either hallucinated or produced vague responses
2. **Mathematical reasoning**: The model could not derive algorithmic complexities or provide mathematical proofs
3. **Depth of explanation**: Some outputs remained shallow, repeating surface-level analogies without deeper technical accuracy
4. **Repetition**: The model sometimes repeated phrases or entire sentences, a known limitation of small language models with limited training data

### Key Insights

1. **Data quality matters more than quantity**: 25 well-written, consistently-formatted samples were enough to change model behavior meaningfully. This shows that for fine-tuning, carefully crafted data outweighs massive but noisy datasets.

2. **LoRA is remarkably efficient**: By updating only 0.4% of model parameters, we achieved noticeable behavioral change. This validates LoRA as a practical approach for domain adaptation, especially with limited compute.

3. **Fine-tuning teaches format, not facts**: The model learned HOW to explain (style, structure, format) better than WHAT to explain (factual content). This is because LoRA primarily adjusts attention patterns, not the model stored knowledge.

4. **Human input is irreplaceable**: The quality of explanations in the training data directly determined the quality of model outputs. Templates from AI-generated content would lack the genuine understanding and creative analogies that come from human experience.

5. **Small models have hard limits**: DistilGPT2 at 82M parameters simply cannot match the reasoning depth of larger models. Fine-tuning improved its output style but could not overcome fundamental capacity limitations.

6. **Fine-tuning is conceptual, not just technical**: Understanding WHY certain choices were made (LoRA vs full fine-tuning, teaching style consistency, targeted data augmentation) is more important than knowing the syntax of the training script. The analytical thinking behind each decision is what makes fine-tuning effective.

### Final Thought
This lab demonstrated that fine-tuning is not just about running code — it requires thoughtful dataset design, understanding model limitations, and iterative improvement. The gap between a generic LLM and a domain-specific assistant is bridged not by more parameters, but by better data and deliberate choices.
